```python
import cv2 as cv
from google.colab.patches import cv2_imshow #tampilanGambar
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/Design Portfolio/Logo/1.png')

plt.imshow(img)
```

```python
img2 = img
img3 = cv.cvtColor(img,cv.COLOR_BGR2RGB) #konversi bgr ke rgb
plt.imshow(img3)
```

```python
img_gray = cv.imread('/content/drive/MyDrive/Design Portfolio/Logo/1.png',cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
plt.imshow(img_gray, cmap='magma')
```

```python
img4 = cv.resize(cv.cvtColor(img,cv.COLOR_BGR2RGB),(512,1024))
plt.imshow(img4)
```

```python
img5 = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)
# Tampil ukuran lebih besar dan terbalik
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img5)
```

```python
black_img = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```
